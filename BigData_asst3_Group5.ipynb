Download transportation trip dataset of Chicago (JSON version):
The following code is creating an API endpoint URL for the City of Chicago's data portal and specifying two query parameters:
Together, these parameters create a query that retrieves a subset of transportation trip data from the City of Chicago's data portal.


```python
!pip install requests

import requests
import json


# Set up the API endpoint and parameters
endpoint = "https://data.cityofchicago.org/resource/wrvz-psew.json"
params = {}
```

The code is sending a GET request to the API endpoint URL specified in the endpoint variable, which is the City of Chicago's data portal. The request includes query parameters specified in the params variable.

```python
# Make a request to the API
response = requests.get(endpoint, params=params)
```

This code checks whether the HTTP response status code received from an API call is equal to 200, which typically indicates that the request was successful. If the response code is indeed 200, the code extracts the JSON data from the response using the .json() method of the response object.

```python
# Check the response status code
if response.status_code == 200:
    # Extract the JSON data from the response
    data = response.json()

    # Save the JSON data to a file
    with open("data.json", "w") as f:
        json.dump(data, f)
        print(data)
```

## Filter data based on Date/Time Range

This code sends a GET request to an API endpoint at the URL specified by the url variable. The endpoint includes a query parameter $where that filters the data to include only records with a trip_start_timestamp between January 1, 2022, and January 31, 2022. The requests.get() function is used to send the request and the response object is stored in the response variable.

```python
url = "https://data.cityofchicago.org/resource/wrvz-psew.json?$where=trip_start_timestamp BETWEEN '2022-01-01T00:00:00' AND '2022-01-31T00:00:00'"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    with open("data.json", "w") as f:
        json.dump(data, f)
else:
    print(f"Request failed with status code {response.status_code}")
```

## Filter data based on Data Type and Date/Time Range

This code repeatedly sends a GET request to an API endpoint that returns CSV data. The endpoint URL includes a query parameter $where that filters the data to include only records with a trip_start_timestamp between January 1, 2022, and January 31, 2022. The loop runs for 100 iterations.
Each time a response is received, the CSV data is saved to a file named "Chicago.csv" in append mode using the open(), write(), and close() functions.
After the loop completes, the last response text is printed to the console using the print() function.

```python
import requests
for i in range(100):
    url = "https://data.cityofchicago.org/resource/wrvz-psew.csv?$where=trip_start_timestamp BETWEEN '2022-01-01T00:00:00' AND '2022-01-31T00:00:00'"
    response = requests.get(url)
    
    data2=response.text
    file1 = open('Chicago.csv', 'a')
    file1.write(data2)
    file1.close()

print(data2)
```

## Filter data based on Geographical Region

This code sends a GET request to an API endpoint that returns JSON data. The URL specifies a query that filters the data to include only records with a pickup_latitude between 41.8748 and 41.8854, and a pickup_longitude between -87.6424 and -87.6226. Finally, the data3 variable is returned, which contains the filtered JSON data from the API endpoint.

```python
url = "https://data.cityofchicago.org/resource/wrvz-psew.json?$where=pickup_latitude between 41.8748 and 41.8854 and pickup_longitude between -87.6424 and -87.6226"

response = requests.get(url)

data3 = response.json()
data3
```
